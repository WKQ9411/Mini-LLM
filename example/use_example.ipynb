1. 通过注册进 transformers 的方式使用

```python
import mini_models  # 先导入用于注册自定义的 Mini-LLM 模型
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
model = AutoModelForCausalLM.from_pretrained("WKQ9411/Mini-Qwen3-Next-160M-A100M-SFT")
tokenizer = AutoTokenizer.from_pretrained("WKQ9411/Mini-Qwen3-Next-160M-A100M-SFT")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
messages = [{"role": "user", "content": "你好，你是谁？"}]
formatted_text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
input_ids = tokenizer(formatted_text, return_tensors="pt")["input_ids"].to(model.device)
response = model.generate(input_ids, max_new_tokens=100)
response = tokenizer.decode(response[0][len(input_ids[0]):], skip_special_tokens=True)
print(response)
```
